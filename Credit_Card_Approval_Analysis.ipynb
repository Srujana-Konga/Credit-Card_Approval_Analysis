## Univariate Analysis

### 1. Categorical Analysis

```python
## Distribution of Gender
sns.countplot(x='Gender',data = df)
plt.title('Gender Distribution')
plt.show()
```

```python
marital_status_values = df['Marital Status'].value_counts()
# Distribution for marital Status
plt.figure(figsize=(7,7))
plt.pie(marital_status_values,labels = marital_status_values.index,autopct = '%1.1f%%',startangle=90)
plt.title('Marital Status Distribution')
plt.show()
```

```python
education_value_counts = df['Education'].value_counts()
# Distribution of Education type in scatter plot
plt.figure(figsize=(12,5))
plt.scatter(education_value_counts.index,education_value_counts.values,color = 'b', s = 100)
plt.title('Distribution of Education Type')
plt.show()
```

```python
# Distribution by Housing Type
sns.countplot(y='Housing Type',data = df)
plt.title('Housing Type Distribution')
plt.xticks(rotation = 45)
plt.show()
```

### 2. Numerical Variables

```python
# Income Distribution
sns.histplot(df['Income'],kde = True)
plt.title('Income Distribution')
plt.show()
```

```python
sns.histplot(df['No. of Children'], kde=True)
plt.title('Number of Children Distribution')
plt.show()
```

```python
sns.boxplot(x=df['Income'])
plt.title('Income Boxplot')
plt.show()
```

```python
# Age distribution by violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['Age'], color='skyblue', inner='quartile')  # 'inner' controls the representation of the quartiles
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
sns.countplot(x='Flag Work Phone', data=df)
plt.title('Flag Mobile Distribution')
plt.show()
```

## Bivariate Analysis

```python
sns.heatmap(df[['Income','No. of Children','Age']].corr(),annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education', y='Income', data=df)
plt.title('Income Distribution by Education Level')
plt.xlabel('Education')
plt.xticks(rotation=45)  
plt.show()
```
